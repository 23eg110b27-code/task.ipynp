```python
#1 .Load the dataset and display shape, columns and data types.
import pandas as pd
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\techwork(3)(2)\messy_employee_dataset_1000_rows.csv")
df
```

```python
#2. calculate percentage of missing values for each column
df.isnull().sum()
df_missing = (df.isnull().sum() / len(df)) * 100


print(df_missing)
```

```python
#3.
df.apply(lambda col: col.map(type).nunique())
```

```python
#4.clean user id by replacing empty strings with nan and converting into uppercase
import numpy as np

df['user_id'] = (
    df['user_id']
    .replace('', np.nan)
    .str.upper()
)
df
```

```python
df['user_id'].head()
```

```python
#5.convert age to numeric and replace invalid values(negative or >100) with nan
import numpy as np
# convert to numeric (invalid parsing → NaN)
df['age'] = pd.to_numeric(df['age'], errors='coerce')
# replace negative ages or ages > 100 with NaN
df.loc[(df['age'] < 0) | (df['age'] > 100), 'age'] = np.nan
```

```python
df['age'].describe()
```

```python
#6.clean salary by removing currency symbols and converting it to numeric
import numpy as np
df['salary'] = (
    df['salary']
    .replace(r'[^\d.]', '', regex=True)  # remove currency symbols & commas
)

df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
```

```python
df['salary'].head()
```

```python
#7.convert join date into datetime format handling multiple date formats
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce', infer_datetime_format=True)
df["joining_year"] = df["join_date"].dt.year
df['join_date']
```

```python
#8.extract joining year from the cleaned join_date column
df["joining_year"] = df["join_date"].dt.year
df["joining_year"]
```

```python
#9 standardize department values by trimming spaces and converting to uppercase
df['department'] = df['department'].str.strip().str.upper()
df['department']
```

```python
#10 normalize city names to a consistent case and spelling
df['city'] = df['city'].str.strip().str.title()
df['city']
```

```python
#11 convert performance score into numeric by mapping text values to numbers
performance_mapping = {
    'Excellent': 3,
    'Good': 2,
    'Poor': 1
}

df['performance_score'] = df['performance_score'].map(performance_mapping)
df['performance_score'] = pd.to_numeric(df['performance_score'], errors='coerce')
df['performance_score']
```

```python
# 12 clean experience years by converting text values to numeric and removing negatives
df['experience_years'] = pd.to_numeric(df['experience_years'], errors='coerce')
df.loc[df['experience_years'] < 0, 'experience_years'] = pd.NA
df['experience_years']
```

```python
# 13 convert is_active values (yes/no/true/false) into boolean format
df['is_active'] = df['is_active'].str.strip().str.lower()
df['is_active'] = df['is_active'].map({'yes': True, 'true': True, 'no': False, 'false': False})
df['is_active']
```

```python
#14 identify and remove duplicate records based on user_id
df = df.drop_duplicates(subset=['user_id'])
df
```

```python
#15 create a clean final dataframe containing only valid, standardized records
clean_df = df.dropna(subset=['user_id', 'age', 'salary', 'join_date', 'department', 'city', 'performance_score', 'experience_years', 'is_active'])
clean_df.reset_index(drop=True, inplace=True)
clean_df
clean_df.info()
clean_df.head()
```
